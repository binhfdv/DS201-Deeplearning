```
def sentence_generator(s):
    s = s.split('\n')
    result = []
    for line in s:
        line = line.split('.')
        for l in line:
            result.append(l)
    fn_result = []
    for _ in result:
        if _!='':
            fn_result.append(_)
    return fn_result
```

```
sentence_generator(df.loc[53, 'label'])
```

```
df.label = df.label.apply(sentence_generator)
```

```
def punctuation_clean(s):
  s = s.strip()
  s = s.lower()
  for i in s:
    if i in string.punctuation+'\n':
      s = s.replace(i, ' ')
  result = ''.join([i for i in s if not i.isdigit()])

  return ' '.join(result.split())

def remove_nonVN(s):
  slist = s.split()
  for i in range(len(slist)):
    if slist[i] not in vn_words:
      s = s.replace(slist[i], ' ')
  return ' '.join(s.split())

# Xóa dấu tiếng Việt
def remove_vn_accent(s):
  s = unidecode(s)
  return s
```
